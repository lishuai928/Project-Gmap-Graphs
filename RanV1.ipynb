```python
import pprint
pp = pprint.PrettyPrinter(indent=4)

# Palo alto geo code.
city = 'Palo Alto, Californiat'
location = '37.4418834,-122.1430195'
northeast = '37.477838, -122.0688679'
southwest = '37.2853458, -122.202476'
type = 'restaurant'

cities = []
place_ids = []
names = []
types = []
vicnities = []
locations = []

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "location": location,
    "radius": 4000, # hard-code distance
#     "rankby": "distance",
    "type": type,
#     "locationbias": 'rectangle:{0}|{1}'.format(southwest, northeast),
    "key": gkey,
}

# extract results
try:
    response = requests.get(base_url, params=params).json()
    if response['status'] != 'OK':
        print('Not OK response: ' + response['status'])
        exit
    
    results = response['results']
#     pp.pprint(results)
    for result in results:
        type = result['types']
        # Skip locality place, e.g. Palo Alto.
        if ('locality' in type):
            continue
        
        cities.append(city)
        place_ids.append(result['place_id'])
        names.append(result['name'])
        types.append(result['types'])
        vicnities.append(result['vicinity'])
        locations.append(to_lat_lng(result['geometry']['location']))
except:
#     error = sys.exc_info()[0]
    print('Exception')

place_df = pd.DataFrame({
    'City': cities,
    'Place ID': place_ids,
    'Name': names,
    'Types': types,
    'Address': vicnities,
    'LatLng': locations,
})
place_df
    
```
